12. Escribe una función que dado un tweet en formato de cadena de caracteres devuelve una lista con los hashtags incluidos.

```python
def hashes(string):
    '''
    Function that takes a string, finds words begining with # and returns a list with all the words that it detected using comprehension lists.

    '''
    string = string.split()
    return [x for x in string if x[0] == '#']
    
b = hashes(chs_tweet)
print(b)
```

14. El método `str.isalnum()` devuelve True si un caracter contiene un número o una letra. Este puede ser empleado para eliminar los caracteres especiales en las palabras `all_tweets_list`. Itera sobre cada palabra y extrae la parte que no contiene signos de puntuación. Crea una nueva lista `words_alnum` con las palabras sin signos de puntuación.

```python
words_alnum = [x for x in all_tweets_list if x.isalnum() == True]
words_alnum_hash = [x for x in all_tweets_list if x.isalnum() == True or x[0] == '#' and x[-1] != ',' and x[-1] != '.']
```

15. Las *stop words*  son palabras comunes en un lenguaje. Si deseamos estudiar las palabras que aparecen en los tweets para identificar temas de interés debemos eliminar las *stop words*. La variable `stopwords` contiene una lista con las palabras vacías del español obtenidas a partir de una paquetería externa. Crea una lista de palabras (`words_cleaned`) que no contenga este tipo de palabras.

```python
stopwords = get_stop_words('spanish')
```

```python
print(stopwords)
```

```python
words_cleaned = [x for x in words_alnum if x not in stopwords]
words_cleaned_hash = [x for x in words_alnum_hash if x not in stopwords]
```

16. Crea un diccionario donde la clave sea cada palabra y su valor las veces que la palabra aparece en el diccionario.

```python
word_freq = {}
for x in words_cleaned:
    word_freq.update({x:words_cleaned.count(x)})
word_freq_hash = {}
for x in words_cleaned_hash:
    word_freq_hash.update({x:words_cleaned_hash.count(x)})
word_freq_hash
```

Ejecute la siguiente línea de código para obtener la nube de palabras de los tweets analizados.

```python
wordcloud = WordCloud(background_color='white',
                      width=720,
                      height=480
                      ).generate_from_frequencies(word_freq)

plt.figure(figsize=(9,6))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
wordcloud2 = WordCloud(background_color='white',
                      width=720,
                      height=480
                      ).generate_from_frequencies(word_freq_hash)

plt.figure(figsize=(9,6))
plt.imshow(wordcloud2)
plt.axis('off')
plt.show()
```

```python
docs = ['the cat is under the table','laura eats apples', 'the dog is under the table', 'paco goes to school'
]
```

```python
index = {}
for i, doc in enumerate(docs):
    for word in doc.split():
        if word not in index:
            index[word] = [i]
        else:
            index[word].append(i)
```

```python
index = {}
for i, doc in enumerate(docs):
    for word in doc.split():
        if word not in index:
            index[word] = {i}
        else:
            index[word].add(i)
```
